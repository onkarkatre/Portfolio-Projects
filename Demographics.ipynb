```python
## To start the browser
from time import sleep, strftime
from datetime import date
import time
from random import randint
from bs4 import BeautifulSoup
import requests
import pandas as pd
import urllib.parse
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pyautogui
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.remote.webelement import WebElement


driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.get("https://geoiq.io/places/400001---Mumbai-G.P.O./z3h9GnmwFN")

headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.111 Safari/537.36'}

pins = []
pin_and_name = []
area = []
population = []
population_density = []
male_population = []
female_population = []
nearest_airport_and_distance = []
nearest_railwaystation_and_distance = []
Top_brands = []
major_commercial_estate = []
known_for = []
full_url = []
commercial = []
today = date.today()

codes = ["400091", "400092", "400093", "400001", "400020"]

for c in codes:
# Opening the pincode box
    pincode_box = driver.find_element(By.CLASS_NAME, 'my-4 ').click()
    sleep(5)

    pyautogui.typewrite(c)
    sleep(3)


    screenWidth, screenHeight = pyautogui.size() # Get the size of the primary monitor.
    currentMouseX, currentMouseY = pyautogui.position() # Get the XY position of the mouse.


    pyautogui.moveTo(88, 330) # Move the mouse to XY coordinates.

    pyautogui.doubleClick()  
    sleep(10)

    current_url = driver.current_url

    new_url = str(current_url)


    url = new_url 

    page = requests.get(url, headers=headers)

    soup = BeautifulSoup(page.content, 'html.parser')
    url

    results = soup.find_all('div', {'col-span-4'})


    for result in results:
        try:
            pins.append(c)
        except:
            pins.append('NA')    

        #try:        
            #seller_name.append(result.find('input', {'name':'data-sellername'}).get('value'))
        #except:
            #seller_name.append('NA')    

        try:        
            commercial.append(result.find('section', {'id':'seo-commercial-overview'}).get_text())
        except:
            commercial.append('NA')

        try:        
            area.append(result.find('section', {'id':'seo-demographics-and-more'}).get_text())
        except:
            area.append('NA')

        try:        
            pin_and_name.append(result.find('h1', {'class':'mt-0 mb-2 text-5xl font-bold'}).get_text())
        except:
            pin_and_name.append('NA')


        full_url.append(url)   


    data = pd.DataFrame({'Date' : today.strftime("%d/%m/%Y"), 'Pincode' : pins, 'Pin_and_name' : pin_and_name, 'full_url' : full_url, 'area' : area, 'commercial' : commercial})

    pyautogui.moveTo(88, 290) # Move the mouse to XY coordinates.

    pyautogui.doubleClick()  
    sleep(3)
    pyautogui.press('delete')
    sleep(3)    

driver.quit()

data
```

```python
data.to_csv('demographics_dataset.csv', index=False)
```
